```python
import math
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime
import requests
import json
from pandas import json_normalize
```

```python
## CODE TO GET DATA
url = "https://v1.handball.api-sports.io/games?league=39&season=2020"

payload={}
headers = {
    'x-rapidapi-key': 'API_KEY',
    'x-rapidapi-host': 'v1.handball.api-sports.io'
    }

response = requests.request("GET", url, headers=headers, data=payload)
response_text = response.text
response_json = json.loads(response_text)

with open("2020.json", "w") as file:
    json.dump(response_json, file)
```

```python
json_list = ['2012.json','2013.json','2014.json','2015.json','2016.json','2017.json','2018.json','2021.json','2022.json']

combined_df = pd.DataFrame()

for ele in json_list:
    with open(ele) as f:
        data = json.load(f)
    df = pd.json_normalize(data)

combined_df.append(df)

combined_df
```

```python
import json
import pandas as pd
from pandas import json_normalize

normalized_data = []

def process_json(file_path):
    with open(file_path) as f:
        data = json.load(f)
    normalized_data.append(json_normalize(data))


for file_path in json_list:
    process_json(file_path)
```

```python
combined_data = pd.concat(normalized_data, ignore_index=True)
```
